```python
import numpy as np

print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ —Å—Ç–æ–ª–±—Ü–µ days_employed', len(df[df['days_employed'].isna()]))
print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ —Å—Ç–æ–ª–±—Ü–µ total_income', len(df[df['total_income'].isna()]))
print('---')
total_income_mean = df.groupby('income_type')['total_income'].mean()
total_income_median = df.groupby('income_type')['total_income'].median()
print('–°—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è', total_income_mean)
print('---')
print('–ú–µ–¥–∏–∞–Ω–∞', total_income_median)

#–ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ —Å—Ç–æ–ª–±—Ü–µ "total_income" –∑–Ω–∞—á–µ–Ω–∏–µ–º –º–µ–¥–∏–∞–Ω—ã
def insert_values_for_income_type(row):
    
    types = row['income_type']
    income = row['total_income']
    
    if types == '—Å–æ—Ç—Ä—É–¥–Ω–∏–∫':
        if np.isnan(income) == True:
            return total_income_median[6]
        else:
            return income
        
    if types == '–∫–æ–º–ø–∞–Ω—å–æ–Ω':
        if np.isnan(income) == True:
            return total_income_median[3]
        else:
            return income
    
    if types == '–ø–µ–Ω—Å–∏–æ–Ω–µ—Ä':
        if np.isnan(income) == True:
            return total_income_median[4]
        else:
            return income
    
    if types == '–≥–æ—Å—Å–ª—É–∂–∞—â–∏–π':
        if np.isnan(income) == True:
            return total_income_median[2]
        else:
            return income
        
    if types == '–ø—Ä–µ–¥–ø—Ä–∏–Ω–∏–º–∞—Ç–µ–ª—å':
        if np.isnan(income) == True:
            return total_income_median[5]
        else:
            return income
        
    if types == '–±–µ–∑—Ä–∞–±–æ—Ç–Ω—ã–π':
        if np.isnan(income) == True:
            return total_income_median[0]
        else:
            return income
        
    if types == '—Å—Ç—É–¥–µ–Ω—Ç':
        if np.isnan(income) == True:
            return total_income_median[7]
        else:
            return income
        
    if types == '–≤ –¥–µ–∫—Ä–µ—Ç–µ':
        if np.isnan(income) == True:
            return total_income_median[1]
        else:
            return income    
        
df['total_income'] = df.apply(insert_values_for_income_type, axis = 1)
display(df.head(15))
```
